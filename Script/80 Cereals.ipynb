# Data Wrangling

```python
#Renaming Columns

df = df.rename(columns = {'mfr' : 'manufacture', 'potass' : 'potassium'})
df
```

### Renaming Variables

```python
# Renaming Variables

result_manufacture = []

for value in df['manufacture']:
    if value == 'N':
        result_manufacture.append('Nabisco')
    elif value == 'Q':
        result_manufacture.append('Quaker')
    elif value == 'A':
        result_manufacture.append('American Home Food Products')
    elif value == 'G':
        result_manufacture.append('General Mills')
    elif value == 'K':
        result_manufacture.append('Kelloggs')
    elif value == 'P':
        result_manufacture.append('Post')
    else:
        result_manufacture.append('Ralston Purina')

```

```python
result_manufacture
```

```python
df['manufacture'] = result_manufacture
```

```python
df['manufacture'].value_counts(dropna = False)
```

### Visualization

```python
# Products barchart

df_manufacture = df['manufacture'].value_counts().sort_values()

fig,ax = plt.subplots(figsize = (10,4), dpi = 90)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.tick_params(bottom=False)
ax.get_yaxis().set_visible(False)

sns.countplot(data=df, x='manufacture')

# add values on top of each bar
ax.bar_label(ax.containers[0])

ax.set_xlabel('', fontsize=13, color = '#ff4800');
fig.suptitle('Cereal Manufactures', fontsize=15, color = '#ff4800')
```

```python
#Renaming Variable values - 2

result_type = []

for value in df['type']:
    if value == 'C':
        result_type.append('Cold')
    else:
        result_type.append('Hot')
```

```python
df['type'] = result_type
```

#### Shelf placement w.r.t nutrients

```python
#Choose ggplot style

plt.style.use('ggplot')

# get figure and axes or subplots
fig, (ax0, ax1, ax2, ax3) = plt.subplots(nrows = 1, ncols = 4, figsize = (15,4))

# prepare data for ax0
x = df['shelf']
y = df['vitamins']

# axes.bar 
ax0.bar(x, y, width = 0.5, align = 'center')
ax0.set(title = 'Vitamins', xlabel = 'shelf', ylabel = 'Amount of Vitamins')

#prepare data for ax1
x = df['shelf']
y = df['carbo']

#axes bar 1
ax1.bar(x, y, width = 0.5, align = 'center')
ax1.set(title = 'Carbo', xlabel = 'shelf', ylabel = 'Amount of Carbohydrates')

#prepare data for ax2
x = df['shelf']
y = df['potassium']

#axes bar 2
ax2.bar(x, y, width = 0.5, align = 'center')
ax2.set(title = 'Potassium', xlabel = 'shelf', ylabel = 'Amount of Potassium')

#Prepare data for ax3
x = df['shelf']
y = df['sugars']

#axes bar 3 
ax3.bar(x, y, width = 0.5, align = 'center')
ax3.set(title = 'sugars', xlabel = 'shelf', ylabel = 'Amount of sugars')

# Title the figure
fig.suptitle('Which Nutrients dominate which Shelf?', fontsize=14, fontweight='bold');
```

Vitamins and Potassium are placed at the Top shelf, whereas Carbo placed at bottom shelf

```python
# get figure and axes or subplots
fig, (ax0, ax1, ax2, ax3) = plt.subplots(nrows = 1, ncols = 4, figsize = (15,4))

# prepare data for ax0
x = df['shelf']
y = df['calories']

# axes.bar 
ax0.bar(x, y, width = 0.5, align = 'center')
ax0.set(title = 'Calories', xlabel = 'shelf', ylabel = 'Amount of Calories')

#prepare data for ax1
x = df['shelf']
y = df['protein']

#axes bar 1
ax1.bar(x, y, width = 0.5, align = 'center')
ax1.set(title = 'Protein', xlabel = 'shelf', ylabel = 'Amount of Proteins')

#prepare data for ax2
x = df['shelf']
y = df['fat']

#axes bar 2
ax2.bar(x, y, width = 0.5, align = 'center')
ax2.set(title = 'Fat', xlabel = 'shelf', ylabel = 'Amount of Fat')

#Prepare data for ax3
x = df['shelf']
y = df['sugars']

#axes bar 3 
ax3.bar(x, y, width = 0.5, align = 'center')
ax3.set(title = 'sugars', xlabel = 'shelf', ylabel = 'Amount of sugars')


# Title the figure
fig.suptitle('Which Nutrients dominate which Shelf?', fontsize=14, fontweight='bold');
```

### Stacked bar Plot

```python
shelf = pd.crosstab(df['manufacture'], df['shelf'],dropna = False)
```

```python
shelf
```

```python
# Get the figure and the axes (or subplots)
fig, ax = plt.subplots(figsize=(10, 5))

# Using dataframe's plot
shelf.plot(kind='bar', ax=ax, fontsize=10, grid=False)

ax.set(xlabel='Manufacturer' , ylabel = 'Frequency',title='Side-by-side barplot')

fig.suptitle('Number of cereals on each shelf (by manufacturer type)', fontsize=14, fontweight='bold');
```
